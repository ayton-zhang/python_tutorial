```python
from timeit import default_timer as timer
my_list = ['a'] * 100000
print(my_list)

start = timer()
my_string = ''
for i in my_list:
    my_string += i
stop = timer()
print(stop-start)
print(my_string)

start = timer()
my_string = ''.join(my_list)
stop = timer()
print(stop-start)
print(my_string)
```
